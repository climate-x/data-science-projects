# Analysing Job Outcomes for Recent Graduates using Pandas Visualization Functionality

In this  project, we'll use the pandas plotting functionality along with the Jupyter notebook interface to explore data quickly using visualizations.
We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by (American Community Survey)[https://www.census.gov/programs-surveys/acs/], which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their (Github repo)[https://github.com/fivethirtyeight/data/tree/master/college-majors].

## Step1: Data Exploration and Cleaning

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
recent_grads=pd.read_csv('recent-grads.csv')
recent_grads.info()
````

````python
print(recent_grads.iloc[0])
print(recent_grads.head())
````

````python
print(recent_grads.tail())
````

````python
raw_data_count=recent_grads.count(axis=0)
print(raw_data_count)
````

````python
recent_grads=recent_grads.dropna()
print(recent_grads.describe())
````

````python
cleaned_data_count=recent_grads.count(axis=0)
print(cleaned_data_count)
````

## Step-2: Analysis using Pandas Visualization Functionality

### Scatter Plots

````python
recent_grads.plot(x='Sample_size',y="Median",kind='scatter',title='Sample_size vs. Median', figsize=(5,10))
````

````python
recent_grads.plot(x='Sample_size',y="Unemployment_rate",kind='scatter',title='Sample_size vs. Unemployment_rate', figsize=(5,10))
````

````python
recent_grads.plot(x='Full_time',y="Median",kind='scatter',title='Full Time Employment vs. Median Income', figsize=(5,10))
````

````python
recent_grads.plot(x='ShareWomen',y="Unemployment_rate",kind='scatter',title='ShareWomen vs. Unemployment_rate', figsize=(5,10))
````

````python
recent_grads.plot(x='Men',y="Median",kind='scatter',title='Men vs. Median', figsize=(5,10))
````

````python
recent_grads.plot(x='Women',y="Median",kind='scatter',title='Women vs. Median', figsize=(5,10))
````

- Sample size does not have correlation with unemployment or median inccome.
- Full time employment has weak correlation with median income.
- Share of women has a weak correlation with unemployment rate
- No of men or women has no correlation with median income.

## Histograms
Instead of creating histograms separately, we will use a loop to create them in 2 rows and 4 columns. We use ```bins=20``` and ```figsize=(14,12)```

````python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(14,12))
for r in range(8):
    ax = fig.add_subplot(2,4,r+1)
    ax = recent_grads[cols[r]].plot(kind='hist',bins=20, rot=40,title=cols[r])
````
