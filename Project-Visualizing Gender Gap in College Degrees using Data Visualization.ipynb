# Project - Visualizing Gender Gap in College Degrees
In this project, we will generate line charts to compare across all degree categories. In the end, we'll explore how to export the final diagram we create as an image file.
***Dataset***
The (Department of Education Statistics)[https://nces.ed.gov/programs/digest/2013menu_tables.asp] releases a data set annually containing the percentage of bachelor's degrees granted to women from 1970 to 2012. The data set is broken up into 17 categories of degrees, with each column as a separate category.

Randal Olson, a data scientist at University of Pennsylvania, has cleaned the data set and made it available on his personal website. You can download the dataset Randal compiled (here)[https://app.dataquest.io/m/147/improving-plot-aesthetics/2/introduction-to-the-data]

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
````

***Following code was used in the lesson "Improving Plot Aesthetics". Since, it used the same dataset, we will use it as reference.***

````python
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
````

## Step-1: Generating Line Plots for all Categories of Degrees
We generate the line charts on the same 3x6 grid with the first row showing `stem_cats`, the second `lib_arts_cats`, and the third `other_cats'
```
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```
- all three of above lists are ordered in descending order by the percentage of degrees awarded to women.
- We will use 3 for loops to generate 17 line charts in a 6x3 grid, using some mathematical ploys in the `range(start,stop,step)` function to get index positions right for `plot()` function.

````python
fig = plt.figure(figsize=(16, 20)) #adjust screen size, should be large enough to accomodate

cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

#first loop for stem_cats
for sp in range(0,18,3): #step=3
    stem_index=int(sp/3) # sp is float,index must be integer not float
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[stem_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[stem_index]], c=cb_orange, label='Men', linewidth=3)
    ax.set_title(stem_cats[stem_index])
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    if stem_index == 0:
        ax.text(2005, 87, 'Men') #adjust positions if required
        ax.text(2002, 8, 'Women')
    elif stem_index == 5:
        ax.text(2005, 58, 'Men')
        ax.text(2001, 42, 'Women') 
#second loop for lib_arts
for sp in range(1,16,3): #step=3, just a little mathematical ploy to get index values right  
    lib_index=int((sp-1)/3) # sp is float,index must be integer not float
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[lib_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[lib_index]], c=cb_orange, label='Men', linewidth=3)
    ax.set_title(lib_arts_cats[lib_index])
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    if lib_index == 0:
        ax.text(2005, 18, 'Men') #adjust positions if required
        ax.text(2002, 72, 'Women')
    elif lib_index == 4:
        ax.text(2005, 52, 'Men')
        ax.text(2001, 46, 'Women') 
#3rd loop for other_cats
for sp in range(2,20,3): #step=3, just a little mathematical ploy to get index values right 
    other_index=int((sp-2)/3) # sp is float,index must be integer not float
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[other_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[other_index]], c=cb_orange, label='Men', linewidth=3)
    ax.set_title(other_cats[other_index])
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[other_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if other_index == 0:
        ax.text(2005, 87, 'Men') #adjust positions if required
        ax.text(2002, 8, 'Women')
    elif other_index == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
````

## Step-4: Adding a Horizontal Line
- We will add a horizontal line at the 50% mark of y-axis in order to allow ease of interpretation by using the `Axes.axhline()` method which needs  the y-axis location for the start of the line:
`ax.axhline(50)`
- We will also give it the `light_gray` color from the (Color Blind 10 palette)[https://tableaufriction.blogspot.com/2012/11/finally-you-can-use-tableau-data-colors.html] having an RGB value of (171, 171, 171). We will set the color using the c parameter.
- We will also set the transparency using the alpha parameter. The value passed in to the alpha parameter must range between 0 and 1.
`ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)`

````python
fig = plt.figure(figsize=(16, 20)) #adjust screen size, should be large enough to accomodate

cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
cb_light_gray=(171/255,171/255,171/255) # chart horizontal line color
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

#first loop for stem_cats
for sp in range(0,18,3): #step=3
    stem_index=int(sp/3) # sp is float,index must be integer not float
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[stem_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[stem_index]], c=cb_orange, label='Men', linewidth=3)
    ax.set_title(stem_cats[stem_index])
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom="off") ## All label ticks off
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100]) #remove all y labels except 0 and 100
    ax.axhline(50,c=cb_light_gray,alpha=0.3) #set the color and transparency of horizontal line
    if stem_index == 0:
        ax.text(2005, 87, 'Men') #adjust positions if required
        ax.text(2002, 8, 'Women')
    elif stem_index == 5:
        ax.text(2005, 58, 'Men')
        ax.text(2001, 42, 'Women')
        ax.tick_params(labelbottom='on') #bottom chart, label on
#second loop for lib_arts
for sp in range(1,16,3): #step=3, just a little mathematical ploy to get index values right  
    lib_index=int((sp-1)/3) # sp is float,index must be integer not float
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[lib_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[lib_index]], c=cb_orange, label='Men', linewidth=3)
    ax.set_title(lib_arts_cats[lib_index])
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom="off") ## All label ticks off
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100]) #remove all y labels except 0 and 100
    ax.axhline(50,c=cb_light_gray,alpha=0.3) #set the color and transparency of horizontal line
    if lib_index == 0:
        ax.text(2005, 18, 'Men') #adjust positions if required
        ax.text(2002, 72, 'Women')
    elif lib_index == 4:
        ax.text(2005, 52, 'Men')
        ax.text(2001, 46, 'Women') 
        ax.tick_params(labelbottom='on') #bottom chart, label on
#3rd loop for other_cats
for sp in range(2,20,3): #step=3, just a little mathematical ploy to get index values right 
    other_index=int((sp-2)/3) # sp is float,index must be integer not float
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[other_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[other_index]], c=cb_orange, label='Men', linewidth=3)
    ax.set_title(other_cats[other_index])
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[other_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom="off") ## All label ticks off
    ax.set_yticks([0,100]) #remove all y labels except 0 and 100
    ax.axhline(50,c=cb_light_gray,alpha=0.3) #set the color and transparency of horizontal line
    if other_index == 0:
        ax.text(2005, 87, 'Men') #adjust positions if required
        ax.text(2002, 8, 'Women')
    elif other_index == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        ax.tick_params(labelbottom='on') #bottom chart, label on
plt.show()
````
